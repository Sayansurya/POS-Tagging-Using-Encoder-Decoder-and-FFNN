```python
from nltk.corpus import brown
import re
from gensim.models import KeyedVectors
```

```python
word2vec = KeyedVectors.load_word2vec_format('word_vectors/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
tagged_corpus = brown.tagged_sents(tagset='universal')
```

```python
x_tagged_words = []
compound_words = []
prt = []
adj = []
x_neighbors = []

unknown_words = {}
total_words = {}
for sentence in tagged_corpus:
    for i, (word, tag) in enumerate(sentence):
        word = word.lower()
        if tag == 'X':
            x_tagged_words.append(word)
            prev_tag = None if i==0 else sentence[i-1][1]
            next_tag = None if i==(len(sentence)-1) else sentence[i+1][1]
            x_neighbors.append((prev_tag, tag, next_tag))
        if re.search('-', word):
            compound_words.append((word, tag))
        if tag == 'PRT':
            prev_word = None if i==0 else sentence[i-1]
            next_word = None if i==(len(sentence)-1) else sentence[i+1]
            prt.append((prev_word, (word,tag), next_word))
        if tag == 'ADJ':
            adj.append(word)
        try:
            word2vec[word]
        except:
            if tag not in unknown_words.keys():
                unknown_words[tag] = 1
            else:
                unknown_words[tag] += 1
        if tag in total_words.keys():
            total_words[tag] += 1
        else:
            total_words[tag] = 1
        
```
